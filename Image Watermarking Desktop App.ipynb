```python
import tkinter as tk
from tkinter import filedialog, messagebox
from PIL import Image, ImageDraw, ImageFont, ImageTk

def open_image():
    global img, img_path, tk_img
    img_path = filedialog.askopenfilename(filetypes=[("Image Files", "*.png;*.jpg;*.jpeg")])
    if img_path:
        img = Image.open(img_path)
        img.thumbnail((400, 400))
        tk_img = ImageTk.PhotoImage(img)
        canvas.create_image(200, 200, image=tk_img)

def add_watermark():
    global img, img_path
    if img_path:
        watermark_text = watermark_entry.get()
        if not watermark_text:
            messagebox.showerror("Error", "Please enter a watermark text.")
            return

        img_with_watermark = img.copy()
        draw = ImageDraw.Draw(img_with_watermark)
        font = ImageFont.truetype("arial.ttf", 30) 
        text_width, text_height = draw.textsize(watermark_text, font=font)
        
        x, y = img_with_watermark.width - text_width - 20, img_with_watermark.height - text_height - 20
        draw.text((x, y), watermark_text, font=font, fill=(255, 255, 255, 128))  
        
        img_with_watermark.thumbnail((400, 400))
        tk_img_watermarked = ImageTk.PhotoImage(img_with_watermark)
        canvas.create_image(200, 200, image=tk_img_watermarked)
        canvas.image = tk_img_watermarked 

        save_button.config(state=tk.NORMAL)
        img_with_watermark.save("watermarked_image.png")

def save_image():
    if img_path:
        save_path = filedialog.asksaveasfilename(defaultextension=".png",
                                                 filetypes=[("PNG files", "*.png"), ("JPEG files", "*.jpg")])
        if save_path:
            img_with_watermark = Image.open("watermarked_image.png")
            img_with_watermark.save(save_path)
            messagebox.showinfo("Success", "Watermarked image saved successfully!")

root = tk.Tk()
root.title("Image Watermarking App")

canvas = tk.Canvas(root, width=400, height=400, bg="gray")
canvas.pack()

btn_frame = tk.Frame(root)
btn_frame.pack(pady=10)

open_button = tk.Button(btn_frame, text="Open Image", command=open_image)
open_button.grid(row=0, column=0, padx=5)

watermark_entry = tk.Entry(btn_frame, width=20)
watermark_entry.grid(row=0, column=1, padx=5)
watermark_entry.insert(0, "Your Watermark")

add_button = tk.Button(btn_frame, text="Add Watermark", command=add_watermark)
add_button.grid(row=0, column=2, padx=5)

save_button = tk.Button(btn_frame, text="Save Image", command=save_image, state=tk.DISABLED)
save_button.grid(row=0, column=3, padx=5)

root.mainloop()
```
